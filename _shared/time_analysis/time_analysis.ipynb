```python
import pandas as pd
# Calculate the average time value for each raw in the column
# Load the CSV file
df = pd.read_csv('data.csv')

# Function to calculate average time
def average_time(time_str):
    times = time_str.split(',')
    total_seconds = sum(int(h) * 3600 + int(m) * 60 + int(s) for h, m, s in (t.split(':') for t in times))
    average_seconds = total_seconds // len(times)
    hours, remainder = divmod(average_seconds, 3600)
    minutes, seconds = divmod(remainder, 60)
    return f"{hours}:{minutes}:{seconds}"

# Calculate average time for each row
df['Average Convergency Time'] = df['Convergency time'].apply(average_time)

# Save the updated DataFrame to a new CSV file
df.to_csv('updated_data.csv', index=False)
```
