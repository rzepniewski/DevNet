```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 20, 100)
```

```python
def funkcja1(x):
    return x**2
def funkcja2(x):
    return 100+200*np.sin(x)
def funkcja3(x):
    return 10*x+1
```

```python
y1 = funkcja1(x)
y2 = funkcja2(x)
y3 = funkcja3(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='x^2', color='blue')
plt.plot(x, y2, label='100+200*sin(x)', color='red')
plt.plot(x, y3, label='10*x+1', color='green')
plt.title('Wykres funkcji x+1, x^2 i sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
